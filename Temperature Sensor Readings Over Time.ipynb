```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned IoT data
df = pd.read_csv("logistics_structured_data.csv")

# Display first few rows to verify data
print(df.head())
```

```python
# Convert timestamp column to datetime
df["timestamp"] = pd.to_datetime(df["timestamp"])
print("\n'timestamp' column converted to datetime format. Verifying column types:")
print(df.dtypes)
```

```python
# Set the visualization style
sns.set(style="whitegrid")
print("\nVisualization style set to 'whitegrid'.")
```

```python
# Create the line plot
plt.figure(figsize=(12, 6))  # Adjust figure size
sns.lineplot(x="timestamp", y="temperature_c", data=df, marker="o")

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add title and labels
plt.title("IoT Temperature Sensor Readings Over Time", fontsize=14)
plt.xlabel("Timestamp", fontsize=12)
plt.ylabel("Temperature (°C)", fontsize=12)

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Save IoT Temperature Sensor Readingsa to a CSV file
df.to_csv("cleaned_iot_data.csv", index=False)
```
